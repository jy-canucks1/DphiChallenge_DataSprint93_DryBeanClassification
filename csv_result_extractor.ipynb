```python
import csv
import numpy as np
import tensorflow as tf
from tensorflow import keras
from google.colab import files
from decimal import Decimal
```

```python
## Load the model
model = keras.models.load_model("bean_classification_final.h5", compile=True)
```

```python
## Get the test data and output the prediction result of the test data to .csv file
def get_test_data(filename):
    results=np.array([])   
    with open(filename) as testing_file:
        csv_reader = csv.reader(testing_file, delimiter=',')
        for row in csv_reader:
            if row[0]=="Area":
              row = False   
            else:
              row[4] = str((Decimal(row[4]) - 1) * 1000)
              row[5] = str((Decimal(row[5])-Decimal(0.75)) * 10000)
              row[8] = str((Decimal(row[8])-Decimal(0.7)) * 10000)
              row[9] = str((1-Decimal(row[9])) * 100000)
              row[10] = str((Decimal(row[10])-Decimal(0.8)) * 1000)
              row[11] = str((Decimal(row[11])-Decimal(0.77)) * 1000)
              row[12] = str(Decimal(row[12]) * 100000)
              row[13] = str(Decimal(row[13]) * 100000)
              row[14] = str(Decimal(row[14]) * 1000)
              row[15] = str((1-Decimal(row[15])) * 1000000)           
              bean_data = row[0:16]
              bean_data = np.array(bean_data).astype('float')
              bean_data = np.transpose(np.expand_dims(bean_data, axis=1))
              prediction = model.predict(bean_data)
              result = np.argmax(prediction, axis=1)
              label = ["CALI", "SIRA", "HOROZ", "BARBUNYA", "DERMASON", "SEKER", "BOMBAY"]
              bean = label[int(result)]           
              results= np.append(results, bean)
    print(results)
    with open('prediction.csv', 'w', newline='') as csvfile:
      fieldname = ['Class']
      pred_writer = csv.DictWriter(csvfile, fieldnames=fieldname)
      pred_writer.writeheader()
      writer = csv.writer(csvfile)
      for item in results:
        writer.writerow([item])
             ## bean_data_as_array = np.array_split(bean_data, 18)


get_test_data('test_dataset.csv') 
```
